```python
import pandas as pd
data={
    "Name":["Chandan","Kartik","Shreyas","Krish","Ayan"],
    "Age":[25,30,50,20,19,],
    "Department":["sales","engineering","hr","cs","cr"],
    "salary":[50000,65000,40000,45000,90000],
}
df=pd.DataFrame(data)
display(df)
```

```python
print("first 3 rows:")
df.head(3)
```

```python
print("column names")
df.columns
```

```python
print("datatypes")
df.dtypes
```

```python
print("describe numerical columns")
df.describe()
```

```python
print("people with salary > 50000")
df[df['salary'] > 50000]
```

Matplotlib:Building Blocks Of Visulization

Line Plot:

```python
import matplotlib.pyplot as plt
years = [2010,2020,2030,2040]
revenue = [100,120,130,110]
plt.plot(years,revenue,marker='o',linestyle='-',color='red')
plt.title("Revenue over the years")
plt.xlabel("years")
plt.ylabel("revenue")
plt.xticks(years)
plt.grid(True)
plt.show()
```

Bar Chart

```python
departments = df['Department'].value_counts()
departments.plot(kind='bar',color='orange')
plt.title('Employee Count By Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

**Histogram**

```python
ages = df['Age']
plt.hist(ages,bins=5,color='pink',edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Pie **Chart**

```python
df.groupby('Department').size().plot(kind='pie',autopct='%1.1f%%',color=['red','blue','yellow'])
plt.title('Department Distribution')
plt.ylabel('')
plt.show()
```

Load Simple **DataSet**

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
print("\ninfomabout the tips dataset:")
tips.info()
print("\nDescription of the tips dataset:")
tips.describe(include='all')
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.countplot(x='day', data=tips,palette='Set2',hue='day',legend=False)
plt.title('Count of Days')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Boxplot of  Tottal Bill By Day')
plt.show()
```
